```python
import matplotlib.pyplot as plt
import numpy as np

# Define the days from March 1st to March 31st
days = np.arange(1, 32)

# Initialize y-axis values
y = np.zeros(31)

# Linear increase from 0 to 100 over 7 days
y[:7] = np.linspace(0, 100, 7)

# Constant at 100 for the next 7 days
y[7:14] = 100

# Linear decrease from 100 to 0 over the next 7 days
y[14:21] = np.linspace(100, 0, 7)

# Plot
plt.figure(figsize=(5, 2))
plt.plot(days, y, linestyle='-', color='purple')
plt.xlabel('Day')
plt.ylabel('Bloom Percentage')
plt.title('Predicted Bloom Percentage Per Day')
plt.grid(False)  # Remove grid lines

# Remove circle dots
plt.xticks([1, 10, 20, 31], ['March 1', 'March 10', 'March 20', 'March 31'])

# Add a red star and label at x = March 10
plt.plot(10, y[9], marker='*', markersize=10, color='red', label='Predicted Peak Bloom')
plt.legend()

plt.show()
```
